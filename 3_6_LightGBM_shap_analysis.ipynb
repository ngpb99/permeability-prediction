```python
import shap
import pandas
import joblib
import pandas as pd
```

```python
#rdkit mol desc
rdkit_train = pd.read_csv('./Caco-2 Permeability Dataset/Filtered Data/Data Split/rdkit_train.csv')
rdkit_test = pd.read_csv('./Caco-2 Permeability Dataset/Filtered Data/Data Split/rdkit_test.csv')
X_train = rdkit_train.drop(columns=['SMILES', 'logPapp'])
y_train = rdkit_train['logPapp']
X_test = rdkit_test.drop(columns=['SMILES', 'logPapp'])
y_test = rdkit_test['logPapp']
```

```python
model = joblib.load('./Trained Models/LightGBM/RDKit Descriptors/rdkit_trained_model_repeat_0.pkl') #All models are the same regardless of repetition
explainer = shap.TreeExplainer(model)
shap_values = explainer(X_test)
```

```python
shap.plots.bar(shap_values)
```

```python
shap.summary_plot(shap_values)
```
