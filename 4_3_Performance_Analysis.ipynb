```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
bar_width = 1
x = pd.Series([i for i in range(1,4)])

r2_combined = [0.3097635102272034, 0.339736499786377, 0.3357178401947022, 0.32836260557174685, 0.336150369644165]
mae_combined = [0.3899074304103851, 0.37091949820518494, 0.3784536898136139, 0.3787256574630737, 0.3762021017074585]
mse_combined = [0.23658493518829346, 0.22614049553871154, 0.22720301747322083, 0.22957584619522095, 0.22719108283519743]
r2_wc = [0.34087845087051394, 0.2477567672729492, 0.2787350952625275, 0.27634137868881226, 0.30735077857971194]
mae_wc = [0.3868132621049881, 0.42048150599002837, 0.40506161749362946, 0.39587391912937164, 0.396783772110939]
mse_wc = [0.22117858082056047, 0.25249533355236053, 0.24204481542110443, 0.24277791529893875, 0.23241590410470964]

y_combined = [np.mean(r2_combined), np.mean(mae_combined), np.mean(mse_combined)]
y_wc = [np.mean(r2_wc), np.mean(mae_wc), np.mean(mse_wc)]
combined_err = [np.std(r2_combined), np.std(mae_combined), np.std(mse_combined)]
wc_err = [np.std(r2_wc), np.std(mae_wc), np.std(mse_wc)]
plt.bar(x - (bar_width/3)/2, y_combined, align='center', width=bar_width/3, alpha=0.5, yerr = combined_err)
plt.bar(x + (bar_width/3)/2, y_wc, align='center', width=bar_width/3, alpha=0.5, yerr = wc_err)
plt.legend(['Combined Dataset', 'Wang & Chen 2020'], loc='upper right')
plt.ylabel('Scores')
plt.title('Performance of GATv2: Combined vs Wang&Chen2020')
plt.xticks([i for i in range(1,4)], ['R2', 'MAE', 'MSE'])
plt.show()
```
