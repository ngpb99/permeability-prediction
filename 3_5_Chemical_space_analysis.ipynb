```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mordred_train = pd.read_csv('./Caco-2 Permeability Dataset/Filtered Data/Data Split/mordred_train.csv')
mordred_test = pd.read_csv('./Caco-2 Permeability Dataset/Filtered Data/Data Split/mordred_test.csv')
```

```python
len(mordred_train)
```

```python
train = pd.read_csv('./Caco-2 Permeability Dataset/Filtered Data/Data Split/train_data.csv')
test = pd.read_csv('./Caco-2 Permeability Dataset/Filtered Data/Data Split/test_data.csv')
```

```python
rdkit_train = pd.read_csv('./Caco-2 Permeability Dataset/Filtered Data/Data Split/rdkit_train.csv')
rdkit_test = pd.read_csv('./Caco-2 Permeability Dataset/Filtered Data/Data Split/rdkit_test.csv')
```

```python
dim_red = pd.read_csv('./Fingerprints and Dim Reduction Dataset/dim_red_full_dataset.csv', index_col=0)
dim_red.reset_index(drop=False, inplace=True)
dim_red.rename(columns={'index':'SMILES'}, inplace=True)
```

```python
mordred_train_smi = pd.DataFrame(mordred_train['SMILES'])
len(mordred_train_smi)
```

```python
mordred_train_dim = pd.merge(mordred_train_smi, dim_red, on='SMILES', how='inner')
len(mordred_train_dim)
```

```python
mordred_test_smi = pd.DataFrame(mordred_test['SMILES'])
train_smi = pd.DataFrame(train['SMILES'])
test_smi = pd.DataFrame(test['SMILES'])
rdkit_train_smi = pd.DataFrame(rdkit_train['SMILES'])
rdkit_test_smi = pd.DataFrame(rdkit_test['SMILES'])
```

```python
mordred_test_dim = pd.merge(mordred_test_smi, dim_red, on='SMILES', how='inner')
train_dim = pd.merge(train_smi, dim_red, on='SMILES', how='inner')
test_dim = pd.merge(test_smi, dim_red, on='SMILES', how='inner')
rdkit_train_dim = pd.merge(rdkit_train_smi, dim_red, on='SMILES', how='inner')
rdkit_test_dim = pd.merge(rdkit_test_smi, dim_red, on='SMILES', how='inner')
```

```python
x = train_dim['0']
y = train_dim['1']

plt.scatter(x, y)
plt.xlabel('UMAP_1')
plt.ylabel('UMAP_2')
plt.xlim(-20, 35)
plt.ylim(-20, 25)
plt.title('Chemical Space Occupied by Original Train Dataset')
plt.show()
```

```python
x = mordred_train_dim['0']
y = mordred_train_dim['1']

plt.scatter(x, y, color='red')
plt.xlabel('UMAP_1')
plt.ylabel('UMAP_2')
plt.xlim(-20, 35)
plt.ylim(-20, 25)
plt.title('Chemical Space Occupied by Mordred Train Dataset')
plt.show()
```

```python
x = rdkit_train_dim['0']
y = rdkit_train_dim['1']

plt.scatter(x, y, color='green')
plt.xlabel('UMAP_1')
plt.ylabel('UMAP_2')
plt.xlim(-20, 35)
plt.ylim(-20, 25)
plt.title('Chemical Space Occupied by RDKit Train Dataset')
plt.show()
```

```python
x = test_dim['0']
y = test_dim['1']

plt.scatter(x, y)
plt.xlabel('UMAP_1')
plt.ylabel('UMAP_2')
plt.xlim(-20, 35)
plt.ylim(-20, 25)
plt.title('Chemical Space Occupied by Original Test Dataset')
plt.show()
```

```python
x = mordred_test_dim['0']
y = mordred_test_dim['1']

plt.scatter(x, y, color='red')
plt.xlabel('UMAP_1')
plt.ylabel('UMAP_2')
plt.xlim(-20, 35)
plt.ylim(-20, 25)
plt.title('Chemical Space Occupied by Mordred Test Dataset')
plt.show()
```

```python
x = rdkit_test_dim['0']
y = rdkit_test_dim['1']

plt.scatter(x, y, color='green')
plt.xlabel('UMAP_1')
plt.ylabel('UMAP_2')
plt.xlim(-20, 35)
plt.ylim(-20, 25)
plt.title('Chemical Space Occupied by RDKit Test Dataset')
plt.show()
```
